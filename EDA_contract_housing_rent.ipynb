```python
#Importing necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Reading the csv files
contract_housing = pd.read_csv("contract housing rent.csv")
```

```python
#Displaying first few rows to understand the data structure
contract_housing.head()
```

```python
#These are the column titles
print(contract_housing.columns)
```

```python
#Dropping columns which are not required
columns_to_drop = [
    'Estimate; With cash rent: - Less than $100',
       'Estimate; With cash rent: - $100 to $149',
       'Estimate; With cash rent: - $150 to $199',
       'Estimate; With cash rent: - $200 to $249',
       'Estimate; With cash rent: - $250 to $299',
       'Estimate; With cash rent: - $300 to $349',
       'Estimate; With cash rent: - $350 to $399',
       'Estimate; With cash rent: - $400 to $449',
       'Estimate; With cash rent: - $450 to $499',
       'Estimate; With cash rent: - $500 to $549',
       'Estimate; With cash rent: - $550 to $599',
       'Estimate; With cash rent: - $600 to $649',
       'Estimate; With cash rent: - $650 to $699',
       'Estimate; With cash rent: - $700 to $749',
       'Estimate; With cash rent: - $750 to $799',
       'Estimate; With cash rent: - $800 to $899',
       'Estimate; With cash rent: - $900 to $999',
       'Estimate; With cash rent: - $1,000 to $1,249',
       'Estimate; With cash rent: - $1,250 to $1,499',
       'Estimate; With cash rent: - $1,500 to $1,999',
       'Estimate; With cash rent: - $2,000 to $2,499',
       'Estimate; With cash rent: - $2,500 to $2,999',
       'Estimate; With cash rent: - $3,000 to $3,499',
       'Estimate; With cash rent: - $3,500 or more', 
]

# Drop the specified columns
contract_housing1 = contract_housing.drop(columns=columns_to_drop)
```

```python
%store -r contract_housing1
```

```python
#The data types of each column
print(contract_housing1.dtypes)
```

```python
#Getting a description of the data
contract_housing1.describe()
```

```python
#Plotting a combined bar graph
rent_columns = ['Low rent', 'Mid rent', 'High rent']
plt.figure(figsize=(15, 8))
ax = contract_housing1[rent_columns].plot(kind='bar', stacked=True, width=0.8)
plt.xlabel('Id')
plt.ylabel('Rent Categories')
plt.title('Rent Categories by Neighborhood')
plt.legend(title='Rent Category')

# Rotate and display every 5th label
for i, label in enumerate(ax.get_xticklabels()):
    if i % 5 != 0:
        label.set_visible(False)

plt.tight_layout()
plt.show()
```

```python
# Plotting a combined bar graph
grouped_data = contract_housing1.reset_index()
plt.figure(figsize=(15, 8))

rent_columns = ['Low rent', 'Mid rent', 'High rent']

for rent_category in rent_columns:
    sns.barplot(x='Neighborhood', 
                y=rent_category, 
                data=grouped_data, 
                label=rent_category)

plt.xlabel('Neighborhood')
plt.ylabel('Count')
plt.title('Comparison of Rent Categories Across Neighborhoods')
plt.legend(loc='upper right')

plt.xticks(rotation=45, ha='right')
plt.xticks(fontsize=8)

plt.tight_layout()
plt.show()
```

```python
#Plotting a bar graph
plt.figure(figsize=(15, 8))
sns.barplot(x='Neighborhood', y='Total rent', data=contract_housing1, ci=None)
plt.xlabel('Neighborhood')
plt.ylabel('Average Total Rent')
plt.title('Average Total Rent Across Neighborhoods')
plt.xticks(rotation=90)
plt.show()
```

```python
#Plotting a correlation matrix
correlation_matrix = contract_housing1[['Estimate; No cash rent', 'Estimate; With cash rent:', 'Total rent', 'Margin of Error; Total:']].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap for Rent Variables')
plt.show()
```

```python
#Checking the number of missing values in each columns
missing_values = contract_housing1.isnull().sum()
print(missing_values)
```
