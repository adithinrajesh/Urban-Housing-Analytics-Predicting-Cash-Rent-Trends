```python
#Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading the csv files
edu = pd.read_csv("Education.csv")
```

```python
#Displaying first few rows to understand the data structure
edu.head()
```

```python
#These are the column titles
print(edu.columns)
```

```python
#The data types of each column
print(edu.dtypes)
```

```python
#Getting a description of the data
edu.describe()
```

```python
# Drop unnecessary columns
columns_to_drop = ["Estimate; Total: - No schooling completed",
"Estimate; Total: - Nursery school",
"Estimate; Total: - Kindergarden",
"Estimate; Total: - 1st grade",
"Estimate; Total: - 2nd grade",
"Estimate; Total: - 3rd grade",
"Estimate; Total: - 4th grade",
"Estimate; Total: - 5th grade",
"Estimate; Total: - 6th grade",
"Estimate; Total: - 7th grade",
"Estimate; Total: - 8th grade",
"Estimate; Total: - 9th grade",
"Estimate; Total: - 10th grade",
"Estimate; Total: - 11th grade",
"Estimate; Total: - 12th grade, no diploma",
"Estimate; Total: - Regular high school diploma",
"Estimate; Total: - GED or alternative credential",
"Estimate; Total: - Some college, less than 1 year",
"Estimate; Total: - Some college, 1 or more years, no degree",
"Estimate; Total: - Associate\'s degree",
"Estimate; Total: - Bachelor\'s degree",
"Estimate; Total: - Master\'s degree",
"Estimate; Total: - Professional school degree",
"Estimate; Total: - Doctorate degree"]
edu_fil= edu.drop(columns=columns_to_drop)
```

```python
%store edu_fil
```

```python
print(edu_fil.columns)
```

```python
# Visualization - Combined Bar Graph for Education Percentages
education_columns = ['No schooling completed percentage', 'Regular high school diploma percentage', 'Associate degree percentage','Bachelors degree percentage', 'Master\'s degree percentage', 'Doctorate degree percentage']
plt.figure(figsize=(15, 8))

custom_labels = ['No sch', 'High Sch', 'Asso Deg', 'Bach Deg', 'Mas Deg', 'Doc Deg']
handles = []
for i, column in enumerate(education_columns):
    sns.barplot(x='Neighborhood', y=column, data=edu_fil, label=custom_labels[i])

# Set y-axis limits
plt.ylim(0, 100)

plt.xlabel('Neighborhood')
plt.ylabel('Percentage')
plt.title('Combined Bar Graph for Education Percentages Across Neighborhoods')

# Set custom labels for the legend
plt.legend(handles=handles, loc='upper right', labels=custom_labels)

plt.xticks(rotation=45, ha='right', fontsize=8)
plt.show()
```

```python
# Swarm Plot
plt.figure(figsize=(15, 8))
sns.swarmplot(x='Neighborhood', y='Estimate education; Total:', data=edu_fil)
plt.title('Distribution of Educational Attainment Across Neighborhoods')
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.show()
```

```python
# Plotting a heatmap for better visualization
correlation_matrix = edu_fil[['Id', 'Estimate education; Total:', 'No schooling completed percentage']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
#Checking the number of missing values in each columns
missing_values = edu_fil.isnull().sum()
print(missing_values)
```

```python
#Performing one hot encoding, feature engineering for neighborhood columns
one_hot_encoded = pd.get_dummies(edu_fil['Neighborhood'], prefix='neighborhood')
edu_fil = pd.concat([edu_fil, one_hot_encoded], axis=1)
```
