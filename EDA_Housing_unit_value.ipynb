```python
#Importing necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Reading the csv files
housing = pd.read_csv("Housing unit value.csv")
```

```python
#Displaying the first few rows to understand the data structure
housing.head()
```

```python
#These are the column titles
print(housing.columns)
```

```python
columns_to_drop = [
"Estimate; Total: - Less than $10,000",
       "Estimate; Total: - $10,000 to $14,999",
       "Estimate; Total: - $15,000 to $19,999",
       "Estimate; Total: - $20,000 to $24,999",
       "Estimate; Total: - $25,000 to $29,999",
       "Estimate; Total: - $30,000 to $34,999",
       "Estimate; Total: - $35,000 to $39,999",
       "Estimate; Total: - $40,000 to $49,999",
       "Estimate; Total: - $50,000 to $59,999",
       "Estimate; Total: - $60,000 to $69,999",
       "Estimate; Total: - $70,000 to $79,999",
       "Estimate; Total: - $80,000 to $89,999",
       "Estimate; Total: - $90,000 to $99,999",
       "Estimate; Total: - $100,000 to $124,999",
       "Estimate; Total: - $125,000 to $149,999",
       "Estimate; Total: - $150,000 to $174,999",
       "Estimate; Total: - $175,000 to $199,999",
       "Estimate; Total: - $200,000 to $249,999",
       "Estimate; Total: - $250,000 to $299,999",
       "Estimate; Total: - $300,000 to $399,999",
       "Estimate; Total: - $400,000 to $499,999",
       "Estimate; Total: - $500,000 to $749,999",
       "Estimate; Total: - $750,000 to $999,999",
       "Estimate; Total: - $1,000,000 to $1,499,999",
       "Estimate; Total: - $1,500,000 to $1,999,999",
       "Estimate; Total: - $2,000,000 or more"
]

data = housing.drop(columns=columns_to_drop)
```

```python
#The data types of each column
print(data.dtypes)
```

```python
#Getting a description of the data
data.describe()
```

```python
#Plotting a combined bar graph
rent_columns = ['Low range', 'Mid range', 'High range']
plt.figure(figsize=(15, 8))
ax = data[rent_columns].plot(kind='bar', stacked=True, width=0.8)
plt.xlabel('Id')
plt.ylabel('Rent Categories')
plt.title('Rent Categories by Neighborhood')
plt.legend(title='Rent Category')

# Rotate and display every 5th label
for i, label in enumerate(ax.get_xticklabels()):
    if i % 5 != 0:
        label.set_visible(False)

plt.tight_layout()
plt.show()
```

```python
#Plotting a combined bar graph
grouped_data = data.reset_index()
plt.figure(figsize=(15, 8))

rent_columns = ['Low range', 'Mid range', 'High range']

# Plotting a combined bar graph
for rent_category in rent_columns:
    sns.barplot(x='Neighborhood', 
                y=rent_category, 
                data=grouped_data, 
                label=rent_category)

plt.xlabel('Neighborhood')
plt.ylabel('Count')
plt.title('Housing unit value estimates')
plt.legend(loc='upper right')

plt.xticks(rotation=45, ha='right')
plt.xticks(fontsize=8)

plt.tight_layout()
plt.show()
```

```python
#Plotting a bar graph
plt.figure(figsize=(15, 8))
sns.barplot(x='Neighborhood', y='Estimate housing unit value; Total:', data=data, ci=None)
plt.xlabel('Neighborhood')
plt.ylabel('Total estimate')
plt.title('Average Total Estimate Across Neighborhoods')
plt.xticks(rotation=90)
plt.show()
```

```python
#Plotting a correlation matrix
correlation_matrix = data[['Id', 'Low range', 'Mid range', 'High range']].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap for Rent Variables')
plt.show()
```

```python
#Checking the number of missing values in each columns
missing_values = data.isnull().sum()
print(missing_values)
```
